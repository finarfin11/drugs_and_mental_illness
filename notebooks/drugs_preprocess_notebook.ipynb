### Metadata

The Global organized crime index dataset has 193 observations of 33 features. It assesses levels of organized crime in 193 countries across the world and also their resilience to organized-criminal activity. Comprising two composite metrics, the Index ranks countries both according to their levels of criminality on a score from 1 to 10 and according to resilience to organized crime from 1 to 10 (lowest to highest levels). The metadata can be viewed [here](../data/guiding_questions.pdf).

The source of this dataset is https://ocindex.net/downloads.
